```python
import requests
from bs4 import BeautifulSoup

# 1) Request the target page
response = requests.get("https://startcoding.pythonanywhere.com/basic")

# 2) Extract raw HTML from the response
html = response.text

# 3) Parse HTML into a BeautifulSoup object
soup = BeautifulSoup(html, "html.parser")

# 4) Find the first element matching the .brand-name selector
soup.select_one(".brand-name")
```

```python
# Get only the visible text content from the selected tag
soup.select_one(".brand-name").text
```

```python
# Check the attributes dictionary of the selected tag
soup.select_one(".brand-name").attrs
```

```python
# Get only the href value (link URL) from attrs
soup.select_one(".brand-name").attrs['href']
```

```python
# Get the target attribute (e.g., _blank)
soup.select_one(".brand-name").attrs['target']
```
