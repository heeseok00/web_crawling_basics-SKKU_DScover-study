```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://startcoding.pythonanywhere.com/basic")
html = response.text
soup = BeautifulSoup(html, "html.parser")
category = soup.select_one(".product-category").text  # select_one: returns only the first matching tag
name = soup.select_one(".product-name").text
link = soup.select_one(".product-name > a").attrs['href']
price = soup.select_one(".product-price").text
# strip(): remove leading/trailing whitespace
price = soup.select_one(".product-price").text.strip()
# replace('old', 'new'): string replacement (currency suffix matches the scraped HTML)
price = soup.select_one(".product-price").text.strip().replace(',', '').replace('\uC6D0', '')
print(category, name, link, price)
```
