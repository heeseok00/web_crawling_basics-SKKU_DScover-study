```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []

for i in range(1, 5):

    response = requests.get(f"https://startcoding.pythonanywhere.com/basic?page={i}")
    html = response.text
    soup = BeautifulSoup(html, "html.parser")
    items = soup.select('.product')

    for item in items:
        category = item.select_one(".product-category").text # select_one: returns only the first matching tag
        name = item.select_one(".product-name").text
        link = item.select_one(".product-name > a").attrs['href']
        price = item.select_one(".product-price").text
        # strip(): remove leading/trailing whitespace
        price = item.select_one(".product-price").text.strip()
        # split on KRW suffix from page, then remove commas
        price = item.select_one(".product-price").text.split('\uC6D0')[0].replace(',', '')
        print(category, name, link, price)
        data.append([category,name,link,price])
```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=['category', 'product_name', 'detail_link', 'price'])
df
```

```python
# Save to Excel
df.to_excel('result.xlsx',index=False)
```
